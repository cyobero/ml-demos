# Linear Regression Tutorial

In this tutorial, we demonstrate linear regression using `sklearn`. Linear regression is a machine learning model whose goal is to predict an output $\hat{y}$ (also known as dependent variable) given one or more set of inputs $x_{i}$ (also known as independent variable(s)).

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import normalize
```

## The Model

Given the mathematical model

$$
Y = \beta_{0} + \beta_{1}X_{i} + \epsilon
$$

where $\beta_{0}$ is the intercept, $\beta_{1}$ is the coefficient of $X$, and $e$ is some random, unobserved error term.

Our goal is to obtain the parameter estimates so that we have

$$
\hat{y} = b_{0} + b_{1}x_{i}
$$

You can visualize this using a scatterplot.

```python
np.random.seed(1234)
x = np.random.normal(5, 1.03, 1000)
y = np.random.uniform(2, 1.5, 1000) + np.random.uniform(10, 5, 1000) * x
m, b = np.polyfit(x, y, 1)

plt.scatter(x, y)
plt.plot(x, m*x + b, c='red')
plt.grid(True)
plt.show()
```

Given a raw set of random data points $(x_{i}, y_{i})$, our goal is to find the "**line of best fit**" (the red line shown above) using a method known as **least squares**.

## Example

First, we import our dataset. The dataset we will be using contains advertising expenditures and sales figures. We will train a regression model to predict an output $\hat{y}$ given one explanatory variable $x_{i}$ (referred to as "simple linear regression").

Let's generate some fake data.

```python
np.random.seed(3)
ad_spend = np.random.normal(100, 30, 1000)
sales = np.random.uniform(1,100, 1000) + np.random.uniform(2, 5, 1000) * ad_spend
```

```python
# normalize our data first
ad_spend = ad_spend / np.linalg.norm(ad_spend)
sales = sales / np.linalg.norm(sales)
```

Let's visualize it.

```python
plt.scatter(ad_spend, sales)
plt.xlabel("Ad Spend (in 000,000$)")
plt.ylabel("Sales (in 000,000$)")
plt.show()
```

## Train-Test Split

We first split out data into training and test sets.

```python
X_train, X_test, y_train, y_test = train_test_split(ad_spend, sales)

# we need to reshape (transpose) our data
X_train = X_train.reshape(-1, 1)
X_test = X_test.reshape(-1, 1)
```

## Model Fitting

```python
model = LinearRegression(normalize=True)
model.fit(X_train, y_train)
```

We can retrieve our parameter estimates as follows...

```python
print("beta_0: {}".format(model.intercept_))
print("beta_1: {}".format(model.coef_[0]))
```

This gives us our fitted equation

$$
\hat{y} = 0.0036 + 0.8627 x_{i}
$$

This means that a one-dollar increase in ad spend results in an $3.5 increase in sales (using the non-normalized data).

## Evaluating Our Model

How do we know if our model is good or not? We need to test it. The test measure we will be using is the **mean squared error**, which can be given as

$$
\text{MSE} = \frac{1}{n} \sum_{i=1}^{N}(y_{i} -\hat{y_{i}})^{2}
$$
